```python
import os
import numpy as np
import fsl.nets as nets
import matplotlib.pyplot as plt
from joblib import Parallel, delayed
from scipy.stats import t
from numpy import sqrt
import shutil

# ========== Paths ==========
component_dir = '/project/data/components/'    # (unused here, but kept for completeness)
png_dir       = '/project/results/melodic_IC_sum.sum/'
output_dir    = '/project/script/nbs_output/'
os.makedirs(output_dir, exist_ok=True)

# ========== Load Connectivity Data ==========
afro  = np.load('/project/data/connectivity/netmats2_afroamerican_paired.npy')
white = np.load('/project/data/connectivity/netmats2_whites_paired.npy')

n_subjects_afro,  n_edges_afro  = afro.shape
n_subjects_white, n_edges_white = white.shape

if n_edges_afro != n_edges_white:
    raise ValueError(f"Afro has {n_edges_afro} edges, White has {n_edges_white} edges—must match.")

n_edges = n_edges_afro
n_subj  = n_subjects_afro + n_subjects_white

# === Infer number of nodes ===
# First try: maybe it's a full N×N flattened
root = np.sqrt(n_edges)
if root.is_integer():
    n_nodes = int(root)
else:
    # Fall back to upper‐triangle formula: n_edges = N(N‒1)/2
    calc_n = (1 + np.sqrt(1 + 8 * n_edges)) / 2
    if not calc_n.is_integer():
        raise ValueError(
            f"n_edges={n_edges} is neither a perfect square nor "
            f"(N*(N-1)/2) for integer N. Cannot infer node count."
        )
    n_nodes = int(calc_n)

print(f"Detected n_nodes = {n_nodes}")

# ========== Build 3D netmats cube: shape (n_nodes, n_nodes, n_subj) ==========
netmats = np.zeros((n_nodes, n_nodes, n_subj), dtype=np.float32)

# If full‐matrix case (n_edges == n_nodes^2), just reshape each row
if n_edges == n_nodes * n_nodes:
    # Fill Afroamerican subjects
    for subj_idx in range(n_subjects_afro):
        vec = afro[subj_idx, :]
        netmats[:, :, subj_idx] = vec.reshape((n_nodes, n_nodes))
    # Fill White subjects
    for subj_idx in range(n_subjects_white):
        vec = white[subj_idx, :]
        netmats[:, :, n_subjects_afro + subj_idx] = vec.reshape((n_nodes, n_nodes))
else:
    # Otherwise, assume upper‐triangle format, so use the old vec_to_mat
    triu_idx = np.triu_indices(n_nodes, k=1)
    def vec_to_mat(uc_vec, n):
        mat = np.zeros((n, n), dtype=np.float32)
        mat[triu_idx] = uc_vec
        mat[(triu_idx[1], triu_idx[0])] = uc_vec
        return mat

    for subj_idx in range(n_subjects_afro):
        netmats[:, :, subj_idx] = vec_to_mat(afro[subj_idx, :], n_nodes)
    for subj_idx in range(n_subjects_white):
        netmats[:, :, n_subjects_afro + subj_idx] = vec_to_mat(white[subj_idx, :], n_nodes)

# Create group labels: 1=Afroamerican, 0=White
groups = np.array([1]*n_subjects_afro + [0]*n_subjects_white)

# ========== Primary threshold ==========
df       = n_subj - 2
p_thresh = 0.001
t_thresh = t.ppf(1 - p_thresh/2, df)
print(f"Using two‐tailed p = {p_thresh} → t_thresh = {t_thresh:.3f}")

# ========== Run NBS (with built‐in permutations) ==========
n_perm = 5000
print("Running NBS with permutations (this may take a while)...")
nbs_stats, nbs_pvals, nbs_clusters, nbs_cluster_sizes = \
    nets.nbs(netmats, groups, thresh=t_thresh, tail='both', n_perm=n_perm)

# Save outputs
np.save(os.path.join(output_dir, 'nbs_stats.npy'),       nbs_stats)
np.save(os.path.join(output_dir, 'nbs_pvals.npy'),       nbs_pvals)
np.save(os.path.join(output_dir, 'nbs_clusters.npy'),    nbs_clusters)
np.save(os.path.join(output_dir, 'nbs_cluster_sizes.npy'), nbs_cluster_sizes)

# ========= Identify significant clusters ==========
signif_clusters = np.where(nbs_pvals < 0.05)[0]  # zero‐based indices
if len(signif_clusters) == 0:
    print("No significant clusters found at p < 0.05.")
else:
    print("Significant clusters (1‐based):", (signif_clusters + 1))
    # Overlay each cluster on the t‐matrix
    for ci in signif_clusters:
        cnum     = ci + 1
        cluster_mask = (nbs_clusters == cnum).astype(int)

        plt.figure(figsize=(6, 6))
        plt.imshow(
            nbs_stats,
            cmap='gray_r',
            vmin=-np.max(np.abs(nbs_stats)),
            vmax= np.max(np.abs(nbs_stats))
        )
        plt.imshow(cluster_mask * nbs_stats, cmap='autumn', alpha=0.7)
        plt.title(f'NBS Cluster {cnum} (p={nbs_pvals[ci]:.4g})')
        plt.colorbar(label='t‐stat (cluster edges)', fraction=0.046, pad=0.04)
        plt.xlabel('Node')
        plt.ylabel('Node')
        plt.xticks(np.arange(0, n_nodes, 5))
        plt.yticks(np.arange(0, n_nodes, 5))
        plt.tight_layout()
        plt.savefig(os.path.join(output_dir, f'nbs_cluster_{cnum}_overlay.png'))
        plt.show()

        # Optional: if you have node coordinates (N×3 array), you can plot a connectome:
        # from nilearn.plotting import plot_connectome
        # coords = load_your_node_coords()  # shape = (n_nodes, 3)
        # cluster_t = cluster_mask * nbs_stats
        # plot_connectome(cluster_t, coords, title=f'Cluster {cnum}')
        # plt.show()

# ========== Copy PNGs for FSLNets GUI (if needed) ==========
if os.path.isdir(png_dir):
    png_out = os.path.join(output_dir, 'node_pngs')
    if not os.path.exists(png_out):
        shutil.copytree(png_dir, png_out)
        print(f"Node PNG images copied to {png_out}")
    else:
        print("Note: 'node_pngs' already exists, skipping copy.")
else:
    print(f"Warning: PNG source '{png_dir}' not found; skipping.")

```

```python
import os
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
import seaborn as sns
import fsl.nets as nets
from scipy.stats import t
from joblib import Parallel, delayed
from nilearn.plotting import plot_connectome

# =============================================================================
# CONFIGURATION
# =============================================================================
component_dir = '/project/data/components/'
output_dir = '/project/script/nbs_output/'
png_dir = '/project/results/melodic_IC_sum.sum/'

os.makedirs(output_dir, exist_ok=True)

# Load connectivity vectors
afro = np.load('/project/data/connectivity/netmats2_afroamerican_paired.npy')
white = np.load('/project/data/connectivity/netmats2_whites_paired.npy')

n_afro, n_edges = afro.shape
n_white, _ = white.shape
n_subj = n_afro + n_white

# Infer number of nodes
calc_n = (1 + np.sqrt(1 + 8 * n_edges)) / 2
if not calc_n.is_integer():
    raise ValueError(f"n_edges={n_edges} is not valid; cannot infer node count.")
n_nodes = int(calc_n)

# Build 3D netmats cube: shape (n_nodes, n_nodes, n_subj)
triu_idx = np.triu_indices(n_nodes, k=1)
netmats = np.zeros((n_nodes, n_nodes, n_subj), dtype=np.float32)

# Fill Afroamerican
netmats[triu_idx[0], triu_idx[1], :n_afro] = afro.T
netmats[triu_idx[1], triu_idx[0], :n_afro] = afro.T

# Fill White
netmats[triu_idx[0], triu_idx[1], n_afro:] = white.T
netmats[triu_idx[1], triu_idx[0], n_afro:] = white.T

groups = np.array([1]*n_afro + [0]*n_white)

# =============================================================================
# PRIMARY THRESHOLD
# =============================================================================
df = n_subj - 2
p_thresh = 0.001
t_thresh = t.ppf(1 - p_thresh/2, df)
print(f"Using two-tailed p={p_thresh:.3f} → t_thresh = {t_thresh:.3f}")

# =============================================================================
# RUN NBS (with internal permutations)
# =============================================================================
n_perm = 5000
print("Running NBS with permutations... this may take a while.")
nbs_stats, nbs_pvals, nbs_clusters, nbs_cluster_sizes = \
    nets.nbs(netmats, groups, thresh=t_thresh, tail='both', n_perm=n_perm)

# Save raw outputs
np.save(os.path.join(output_dir, 'nbs_stats.npy'), nbs_stats)
np.save(os.path.join(output_dir, 'nbs_pvals.npy'), nbs_pvals)
np.save(os.path.join(output_dir, 'nbs_clusters.npy'), nbs_clusters)
np.save(os.path.join(output_dir, 'nbs_cluster_sizes.npy'), nbs_cluster_sizes)

# =============================================================================
# IDENTIFY SIGNIFICANT CLUSTERS
# =============================================================================
signif_clusters = np.where(nbs_pvals < 0.05)[0]
print("Significant clusters (p<0.05):", signif_clusters + 1)

if len(signif_clusters) == 0:
    print("No significant clusters found.")
else:
    # For each cluster, plot adjacency heatmap overlay
    for idx in signif_clusters:
        cnum = idx + 1
        mask = (nbs_clusters == cnum).astype(int)
        plt.figure(figsize=(6, 6))
        plt.imshow(nbs_stats, cmap='gray_r', vmin=-np.max(np.abs(nbs_stats)), vmax=np.max(np.abs(nbs_stats)))
        plt.imshow(mask * nbs_stats, cmap='autumn', alpha=0.7)
        plt.title(f'NBS Cluster {cnum} (p={nbs_pvals[idx]:.4g})')
        plt.colorbar(label='t-stat value (cluster edges)', fraction=0.046, pad=0.04)
        plt.xlabel('Node')
        plt.ylabel('Node')
        plt.xticks(np.arange(0, n_nodes, 5))
        plt.yticks(np.arange(0, n_nodes, 5))
        plt.tight_layout()
        plt.savefig(os.path.join(output_dir, f'nbs_cluster_{cnum}_overlay.png'))
        plt.show()

        # Optionally: plot as a connectome if you have node coordinates
        # coords = load_your_node_coords(...)  # shape (n_nodes, 3)
        # cluster_masked_t = mask * nbs_stats
        # plot_connectome(cluster_masked_t, coords, title=f'Cluster {cnum}')
        # plt.show()

# =============================================================================
# OPTIONAL: COPY NODE PNGs for FSLNets GUI
# =============================================================================
if os.path.isdir(png_dir):
    png_out = os.path.join(output_dir, 'node_pngs')
    if not os.path.exists(png_out):
        shutil.copytree(png_dir, png_out)
    print(f"Copied node PNGs to {png_out}")
else:
    print(f"Warning: PNG directory '{png_dir}' not found, skipping copy.")
```
